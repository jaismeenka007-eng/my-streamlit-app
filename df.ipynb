```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('diabetes_dataset.csv')
df
```

```python
#countplot
sns.countplot(df.head(20),x="diabetes_risk_score",palette="Set2")
plt.xticks(rotation=45)
```

```python
sns.countplot(df,x="alcohol_consumption_per_week",palette="Set2")
```

```python
sns.barplot(df.head(10),x="diabetes_risk_score",y="alcohol_consumption_per_week",hue="gender",palette="Set2")
```

```python
sns.lineplot(df.head(10),x="ethnicity",y="smoking_status",hue="gender",palette="Set1")
```

```python
sns.barplot(df,x="diabetes_risk_score",y="diabetes_stage",hue="gender")
```

```python
sns.scatterplot(df,x="diabetes_stage",y="age",hue="gender",palette="Set1")
```

```python
plt.pie(df['diabetes_stage'].value_counts(),labels=df['diabetes_stage'].unique(),autopct='%1.1f%%',colors=sns.color_palette('Set2'))
```

```python
plt.hist(df['age'],bins=20,color='skyblue',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
sns.heatmap(df[['age','alcohol_consumption_per_week','diabetes_risk_score']].corr(),annot=True,cmap='coolwarm')
```

```python
sns.heatmap(df[['age','physical_activity_minutes_per_week','diabetes_risk_score']].corr(),annot=True)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(df.head(100),x="age",y="diabetes_risk_score",hue="gender",palette="Set1")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df.head(100),x="age",y="diabetes_risk_score",hue="gender",palette="Set1")
plt.title("Diabetes Risk Score by Age and Gender")
plt.xticks(rotation=45)
```

```python
plt.pie(df['ethnicity'].value_counts(),labels=df['ethnicity'].unique(),autopct='%1.1f%%',colors=sns.color_palette('Set1'))
plt.title("Distribution of Ethnicity")
plt.axis('equal')
```
